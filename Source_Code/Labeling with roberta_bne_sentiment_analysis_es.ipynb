# Imports

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import copy
from google.colab import files
from tqdm.notebook import tqdm
```

#Pipeline de procesado

## Subir el csv

```python
uploaded = files.upload()
for filename in uploaded.keys():
  dataframe = pd.read_csv(filename)
```

```python
dataframe.head()
```

## Calcular sentiment análisis

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch.nn.functional as F

tokenizer = AutoTokenizer.from_pretrained("edumunozsala/roberta_bne_sentiment_analysis_es")
model = AutoModelForSequenceClassification.from_pretrained("edumunozsala/roberta_bne_sentiment_analysis_es")

reviews = dataframe['review']

for i, text in tqdm(enumerate(reviews), total=len(reviews)):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512)
    outputs = model(**inputs)
    logits = outputs.logits
    probs = F.softmax(logits, dim=1)
    labels = {0: "negative", 1: "positive"}

    for j, prob in enumerate(probs.squeeze()):
        dataframe.at[i, labels[j]] = round(prob.item() * 100, 2)
```
